```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def inside_unit_circle(x, y):
    return x**2 + y**2 <= 1
```

```python
assert(inside_unit_circle(0, 0))
assert(inside_unit_circle(0.5, 0.5))
assert(not inside_unit_circle(1, 1))
```

```python
population = 150000
x = np.random.uniform(-1, 1, population)
y = np.random.uniform(-1, 1, population)
```

```python
inside = np.array([inside_unit_circle(x[i], y[i]) for i in range(population)])
outside = np.array([not inside_unit_circle(x[i], y[i]) for i in range(population)])
```

```python
# scatter plot - black dots outside, red dots inside
plt.scatter(x[outside], y[outside], color='black')
plt.scatter(x[inside], y[inside], color='red')
# set aspect ratio to 1:1
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
ratio = np.sum(inside) / population
ratio * 4
```
